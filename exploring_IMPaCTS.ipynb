```python
import pandas as pd
```

### Opening IMPaCTS with Pandas

```python
df = pd.read_csv("IMPaCTS.zip", compression="zip")
len(df)
```

### Choosing a Domain

```python
wiki = df[df["domain"] == "wikipedia"]
pa = df[df["domain"] == "public_administration"]
```

```python
wiki
```

```python
pa
```

### Getting all original sentences

```python
pa_og = pa.groupby("original_sentence_idx").nth(0) # gets only the first pair for each original sentence in Public Administration domain
print(f"Number of original sentences in Public Administration domain: {len(pa_og)} / Number of total sentence pairs: {len(pa)}")
```

```python
wiki_og = wiki.groupby("original_sentence_idx").nth(0) # gets only the first pair for each original sentence in Wikipedia domain
print(f"Number of original sentences in Wikipedia domain: {len(wiki_og)} / Number of total sentence pairs: {len(wiki)}")
```

### Extracting a specific original sentence and all its simplification ranked by Read-IT score

```python
target_original_sentence_id = 110992
```

```python
target_pairs = df[df['original_sentence_idx'] == target_original_sentence_id]
```

```python
ordered_pairs = target_pairs.sort_values(by='simplification_all', ascending=False).reset_index(drop=True)
```

```python
for index, row in ordered_pairs.iterrows():
    if index == 0:
        print(f"Original Sentence: <<{row['original_text']}>>\nRead-IT Score: {row['original_all']}\n-----")
    else:
        print(f"Simplified Sentence: <<{row['simplification']}>>\nRead-IT Score: {row['simplification_all']}\n-----")
```

### Streaming IMPaCTS from HuggingFace Hub

IMPaCTS is also on HuggingFace [here](https://huggingface.co/datasets/mpapucci/impacts). It can be loaded using the datasets library, by either downloading the entire datasets, or streaming it directly from HuggingFace.

```python
!pip install datasets
```

```python
from datasets import load_dataset

# Stream a specific domain
dataset = load_dataset("mpapucci/impacts", split="all", streaming=True)
# or: dataset = load_dataset("mpapucci/impacts", split="public_administration", streaming=True)
# or: dataset = load_dataset("mpapucci/impacts", split="wikipedia", streaming=True)

# Iterate through dataset without loading entire thing into memory
for i, example in enumerate(dataset):
    if i >= 5:  # Show first 5 examples
        break
    print(f"\nExample {i+1}:")
    print(f"Domain: {example['domain']}")
    print(f"Original: {example['original_text'][:80]}...")
    print(f"Simplified: {example['simplification'][:80]}...")
```
